```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
```

### 1. Plot a plane f(x,y) = x+y

```python
# Let's plot the graph of the function

# First we define the function 
def func(x,y):
    return x+y
```

```python
# Building the grid in the x-y plane

xMin = -5; xMax = 5; yMin = -6; yMax = 6     # Extrema of the grid
nPoints = 100                                # N. of points in the grid
```

```python
x = np.linspace(xMin,xMax,nPoints)           # Range of x
y = np.linspace(yMin,yMax,nPoints)           # Range of y
X,Y = np.meshgrid(x, y)                      # Coordinates of x and y in the grid (create a 2-matrix with the vector x and y, 
                                             # it is necessary to create all combinations)
Z = func(X, Y)             
#print(x)
#print(y)
#print(Z)
```

```python
# 3-D plot

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z, cmap=cm.viridis)

fig.colorbar(surf, shrink=0.5, aspect=5)
plt.title('$f(x,y)=x+y$')
plt.show()

# Level curves

plt.contour(X,Y,Z,cmap=cm.viridis)
plt.title('Level curves')
plt.show()
```

### 2. Quadratic form: a concave upword paraboloid

##### f(x,y) = 2x^2 + 2xy + 3y^2

```python
# f(x,y) = a*x^2 + b*y^2 + 2*c*xy --> 
# 2x^2 + 2xy + 3y^2 --> a=2 , b=3 , c=2/2
Q = np.array([[2,1],[1,3]])
print(Q)
```

```python
print(np.linalg.det(Q))
# The det Q > 0 and a > 0 then the quadratic form is positive definite and f (x, y) is a concave upword paraboloid
```

```python
# compute the eigenvalues and the eigenvectors
w,v=np.linalg.eig(Q)
print(w)
print(v)
# The 2 eigenvalues are positive, then the quadratic form is positive definite and f (x, y) is a concave upword paraboloid
```

```python
Q@v[:,0]
```

```python
w[0]*v[:,0]
```

```python
# Check the definition Qv = w v
# Pay attention! the first eigenvector is the first column of v

np.allclose(Q@v[:,0],w[0]*v[:,0]) #Returns True if two arrays are element-wise equal within a tolerance
```

```python
np.allclose(Q@v[:,1],w[1]*v[:,1])
```

```python
# Let's plot the graph of the function 

# First we define the function 
def func(x,y):                               
    return 2*x**2+2*x*y+3*y**2
```

### 3. Quadratic form: a concave downword paraboloid

##### f(x,y) = -3x^2 + xy -y^2

```python
Q = np.array([[-3,0.5],[0.5,-1]])
print(Q)
```

```python
print(np.linalg.det(Q))
# The det Q > 0 and a < 0, the quadratic form is negative definite and f (x, y) is a concave downword paraboloid
```

```python
# compute the eigenvalues and the eigenvectors
w,v=np.linalg.eig(Q)
print('Eigenvalue:', w)
print('Eigenvector', v)
# The eigenvalues are negative, then the quadratic form is negative definite and f (x, y) is a concave downword paraboloid
```

```python
# Check the definition 

np.allclose(Q@v[:,0],w[0]*v[:,0])
```

```python
np.allclose(Q@v[:,1],w[1]*v[:,1])
```

```python
x = np.linspace(xMin,xMax,nPoints)           # Range of x
y = np.linspace(yMin,yMax,nPoints)           # Range of y
X,Y = np.meshgrid(x, y)                      # Coordinates of x and y in the grid 
Z = func(X, Y)                               # Values of the function in the points of the grid
```

### 4. Quadratic form: a saddle

#####  f(x,y) = 2x^2 + 6xy -y^2

```python
Q = np.array([[2,3],[3,-1]])
print(Q)
#Q = np.array([[2,0],[0,-1]])
#print(Q)
```

```python
print(np.linalg.det(Q))
# The det Q < 0, the quadratic form is indefinite and f (x, y) is a saddle
```

```python
# compute the eigenvalues and the eigenvectors
w,v=np.linalg.eig(Q)
print('Eigenvalue:', w)
print('Eigenvector', v)
# One eigenvalue is negative and the other is positive, then the quadratic form is indefinite and f (x, y) is a saddle
```

```python
# Check the definition

np.allclose(Q@v[:,0],w[0]*v[:,0])
```

```python
np.allclose(Q@v[:,1],w[1]*v[:,1])
```

### 5. Quadratic form: a valley of minima

##### f(x,y) = 4x^2 + 4xy +y^2

```python
Q = np.array([[4,2],[2,1]])
print(Q)
```

```python
print(np.linalg.det(Q))
# The det Q = 0, the quadratic form is semi-definite (either positive or negative) and
# f(x, y) is a either a valley of minima or a crest of maxima
```

```python
# compute the eigenvalues and the eigenvectors
w,v=np.linalg.eig(Q)
print('Eigenvalue:', w)
print('Eigenvector', v)
# One eigen-value is positive and the other is zero, then the quadratic form is positive semi-definite and
# f(x, y) is a valley of minima
```

```python
# Check the definition

np.allclose(Q@v[:,0],w[0]*v[:,0])
```

```python
np.allclose(Q@v[:,1],w[1]*v[:,1])
```

```python
# Let's plot the graph of the function 

# First we define the function 
def func(x,y):                               
    return 4*x**2+4*x*y+y**2
```

```python
# Building the grid in the x-y plane

xMin = -5; xMax = 5; yMin = -5; yMax = 5     # Extrema of the grid
nPoints = 100                                # N. of points in the grid
```

```python
x = np.linspace(xMin,xMax,nPoints)           # Range of x
y = np.linspace(yMin,yMax,nPoints)           # Range of y
X,Y = np.meshgrid(x, y)                      # Coordinates of x and y in the grid 
Z = func(X, Y)                               # Values of the function in the points of the grid
```

```python
# 3-D plot

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

surf = ax.plot_surface(X, Y, Z, cmap=cm.RdBu)
fig.colorbar(surf)
plt.title('f(x,y) = 4x^2 + 4xy +y^2')
plt.show()

# Level curves 

plt.contour(x,y,Z,cmap=cm.RdBu)
# Let's draw the level curve that contains the eigenvector (the one that has eigenvalue equal to zero)
plt.plot(x,(v[1,1]/v[0,1])*x, color='r')
plt.ylim([-5, 5])
plt.plot(v[0,1], v[1,1], marker="o", markersize=5, markerfacecolor="green")
plt.title('Level curves')
plt.show()

# Along this eigenvector we have the valley of the minima
```

### 6. Quadratic form: a crest of maxima

##### f(x,y) = -4x^2 -4xy -y^2

```python
Q = np.array([[-4,-2],[-2,-1]])
print(Q)
```

```python
print(np.linalg.det(Q))
# The det Q = 0, the quadratic form is semi-definite (either positive or negative) and
# f(x, y) is a either a valley of minima or a crest of maxima
```

```python
# compute the eigenvalues and the eigenvectors
w,v=np.linalg.eig(Q)
print('Eigenvalue:', w)
print('Eigenvector', v)
# One eigenvalue is negative and the other one is zero, then the quadratic form is negative semi-definite and
# f(x, y) is a crest of maxima
```

```python
# Check if the eigen-values and the eigen-vectors are correct

np.allclose(Q@v[:,0],w[0]*v[:,0])
```

```python
np.allclose(Q@v[:,1],w[1]*v[:,1])
```

```python
# Building the grid in the x-y plane

xMin = -5; xMax = 5; yMin = -5; yMax = 5     # Extrema of the grid
nPoints = 100                                # N. of points in the grid
```

```python
x = np.linspace(xMin,xMax,nPoints)           # Range of x
y = np.linspace(yMin,yMax,nPoints)           # Range of y
X,Y = np.meshgrid(x, y)                      # Coordinates of x and y in the grid 
Z = func(X, Y)                               # Values of the function in the points of the grid
```

```python
fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
surf = ax.plot_surface(X, Y, Z,cmap=cm.RdBu)

fig.colorbar(surf)
plt.title('$f(x,y) = -4x^2 -4xy -y^2$')
plt.show()

# Level curves 
plt.contour(x,y,Z,cmap=cm.RdBu)
# Let's draw the level curve that contains the eigenvector (the one that has eigenvalue equal to zero)
plt.plot(x,(v[1,1]/v[0,1])*x, color='b')
plt.ylim([-5, 5])
plt.plot(v[0,1], v[1,1], marker="o", markersize=5, markerfacecolor="green")
plt.title('Level curves')
plt.show()

# Along this eigenvector we have the crest of maxima
```
